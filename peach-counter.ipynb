```python
import os
import random
import shutil

# Paths for train images and labels
train_images_dir = '/kaggle/working/Nectarino-1/train/images'
train_labels_dir = '/kaggle/working/Nectarino-1/train/labels'

# Destination paths for validation images and labels
valid_images_dir = '/kaggle/working/Nectarino-1/valid/images'
valid_labels_dir = '/kaggle/working/Nectarino-1/valid/labels'

# Create validation directories if they don't exist
os.makedirs(valid_images_dir, exist_ok=True)
os.makedirs(valid_labels_dir, exist_ok=True)

# Get list of all image files in the training directory
image_files = [f for f in os.listdir(train_images_dir) if f.endswith(('.jpg', '.png', '.jpeg'))]

# Set the percentage of images to use for validation (e.g., 20%)
valid_split_percentage = 0.01
num_valid = int(len(image_files) * valid_split_percentage)

# Randomly select images for validation
valid_images = random.sample(image_files, num_valid)

# Move selected images and their corresponding label files to the validation directory
for image_file in valid_images:
    # Move image file
    src_image_path = os.path.join(train_images_dir, image_file)
    dst_image_path = os.path.join(valid_images_dir, image_file)
    shutil.move(src_image_path, dst_image_path)

    # Move the corresponding label file (assumes labels have the same filename with .txt extension)
    label_file = image_file.replace(image_file.split('.')[-1], 'txt')
    src_label_path = os.path.join(train_labels_dir, label_file)
    dst_label_path = os.path.join(valid_labels_dir, label_file)

    # Check if label file exists before moving
    if os.path.exists(src_label_path):
        shutil.move(src_label_path, dst_label_path)

print(f"Moved {num_valid} images and their labels to the validation set.")
```
